### Parsing the GFALL Kurucz file ###

Download `gfall.dat`.

```python
!wget -q http://kurucz.harvard.edu/linelists/gfall/gfall.dat
```

Import the `GFALLReader` class and load the data.

```python
from carsus.io.kurucz import GFALLReader

gfall_reader = GFALLReader('gfall.dat')
```

Raw parsed file is stored in the `gfall_raw` attribute.

```python
gfall_reader.gfall_raw
```

Dump the data with the `to_hdf` method.

```python
gfall_reader.to_hdf('test.h5')
```
