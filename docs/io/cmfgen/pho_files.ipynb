# Photoionization cross-sections

```python
from carsus.io.cmfgen import CMFGENPhotoionizationCrossSectionParser
```

Download and extract CMFGEN Atomic Data from http://kookaburra.phyast.pitt.edu/hillier/web/CMFGEN.htm

```python
pho_cross_sections = CMFGENPhotoionizationCrossSectionParser('./atomic/SIL/II/16sep15/phot_nahar_A')
```

File header is stored in the *meta* attribute.

```python
pho_cross_sections.meta
```

Attribute *base* is a list containing a DataFrame per configuration.

```python
type(pho_cross_sections.base)
```

```python
pho_cross_sections.base[0]
```

You can easily dump the data inside an HDF5 file. For example:

```python
pho_cross_sections.to_hdf()
```

will dump all the cross-sections tables and metadata inside the file `phot_nahar_A.h5` located in the same path of the original file.
